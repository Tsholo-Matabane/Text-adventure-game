```python
#Docstring

# Name : Tsholofelo Matabane
# Date : 25 October 2021
# Description : Jumanji Text Adventure Game
"""
DocString:
    
    A) Introduction:
    
Welcome to the Jungle

The game you about to play is based on Jumanji December 2017.This game is 
about four teenagers who found an old game console and decided to play it. 
Upon choosing avatars they get sucked into the  jungle world of Jumanji. Once
in the jungle, they are handed the magicaljewel Jaguar’s Eye that was stolen 
which created a curse in Jumanji.  

The goal is to use they complementary skills return the jewel to the
statue at the shrine to lift the curse that fall in jungle and call out\n 
“Jumanji” to exit (win) the game and return to the real world. To return
the jewel, you will go past three adventurous stages without dying as 
you only have one life in the game.\n 
    
Stage 1: The Mighty Roar 
Stage 2: Transportation Shed 
Stage 3: Jaguar Shrine
    
B) Known Bugs and/or Errors:
    
    
"""
#imported package
import time


#Start by select your avatar_def start
def start():
    print("""
Before you begin, please select your avatar.
1) Dr. Bravestone.
2) Ruby Roundhouse.
    """)
        
    choice = input(prompt = ">")
    choice = choice.casefold()
    #code shows the selected Dr. Bravestone
    if "1" in choice or "d" in choice or "D" in choice:
        print("Welcome to Jumanji Dr. Smolder Bravestone B")
        
        
        choice = input(prompt="""\nWould you like to see your strengths and weaknesses? Yes/No>\n""")
                       
        #Dr. Bravestone character traits
        if "yes" in choice or "y" in choice:
            print("""\nStrength = speed and climbing. Boomerang, fearless,Smoldering Intensity
Weakness = Switchblade.\n""")
            input("""<The fate of Jumanji is in your hands>\n
            Press enter to continue>\n""")
            the_mighty_roar()
        
        #when user doesn't want to see the traits,the user is taken to the next stage
        elif "no" in choice or "n" in choice:
            print("""<The fate of Jumanji is in your hands>\n
            Press enter to continue>\n""")
            the_mighty_roar()
        
    #code shows the selected Ruby Roundhouse         
    elif "2" in choice or "Ruby Roundhouse" in choice or "Ruby" or "Roundhouse" in choice :
        print("Welcome to Jumanji Ruby Roundhouse - The Killer of Men ")
        choice = input(prompt="""\nWould you like to see your strengths and 
        weaknesses? Yes/No>\n""")
        choice = choice.lower()
        
        #Ruby Roundhouse character traits
        if "yes" in choice or "y" in choice:
            print("""\nStrength = The killer of men”, Karate, Tai ‘Chi, Aikido, 
            Dance Fighting and Nun chucks  
            Weakness = Venom.\n""")
            input("""<The fate of Jumanji is in your hands>\n
            Press enter to continue>\n""")
            the_mighty_roar()
            
        #when user doesn't want to see the traits,the user is taken to the next stage
        elif "no" in choice or "n" in choice:
            print("""<The fate of Jumanji is in your hands>\n
            Press enter to continue>\n""")
            the_mighty_roar()
            
    # identifies incorrect inputs
    else:
        print("Invalid selection. Please re-select your avatar")
        start()
        
        
    

# intoduction of stage_1 
def the_mighty_roar():
    print("""
The Mighty Roar\n 
You are now sucked into the game and you landed in the Deep jungle 
of Jumanji in a transformed body of your selected avatar and met with 
Nigel Billingsley who handed you the jewel and transported you to The 
Mighty Roar.""")

#enables continuation to the next part of the stage
    print(input("Press enter to continue>\n"))

#user is attacked and must select one strength to use 
    print("""
    Middle of the Jungle\n 
    As you are standing in the middle of the jungle perplexed,you hear a loud  
    sound of traditional drumbeats coming from the forest and instantly the you 
    are attacked by Professor Van Pelt's Army of 35 soldiers, armed with 
    motorbikes and automatic machine guns.\n""")

    #user options to respond to the attack
    print("""
    1) Use your speed skill to run for your life.
    2) Attack with your Boomerang.
    3) Use your climbing skill to hide on top of the tree
    """)
    
    # user choice input
    choice = input(">  \n")

    #results of user selecting the first option
    if "1" in choice or "run" in choice or "speed" in choice:
        print("\nOff you go!\n")
        input("<Press enter to continue>\n")
        waterfall_Cliff()
        
    #results of user selecting the second option
    elif "2" in choice or "attack" in choice or "boomerang" in choice:
        print("""\n Attack Attack Attack!!!!""")
        
        #user given three chances to throw the boomerang and everytime there is throw one point is decreased
        boomerang= 2
        while boomerang >= 0:
            input(f"<Press enter to throw your boomerang. You have {boomerang+1} left>\n")
            boomerang -=1
        waterfall_Cliff()
            
    #results of user selecting the third option which will result in a fail
    elif "3" in choice or "climb" in choice or "hide" in choice:
        print("\nYou were shot ontop of the tree.")
        input("You have failed to lift the curse of Jumanji. Game Over!!")
        fail()
    
    # identifies incorrect inputs
    else:
        print("Invalid command. Please try again.")
        input("<Press enter to continue>\n")
        start()



# continuation of stage_1 in a different setting. User given three options to select from
def waterfall_Cliff():
    print("""
Waterfall Cliff
You managed to run safely however you ran towards the end of the highest
waterfall cliff with the deepest river flowing at the bottom of the waterfall.
As you were running the army of soldiers was following you behind.

As you stand there trembling, you are faced with the following options;
1) Jump off the cliff.
2) Run in the opposite direction. 
      """)

    choice = input(">  \n")
    
    if "1" in choice or "jump" in choice or "Jump off the cliff" in choice:
        print("\nYou're off to the river....screaming!!!!\n")
        input("<Press enter to continue>\n")
        transportation_Shed() 
    
    elif "2" in choice or "run" in choice or "Run in the opposite direction" in choice:
        print("""\nAs you were running you got shot. Game Over""")
        fail()

    else:
        print("Invalid entry. Please try again.")
        input("<Press enter to continue>\n")
        waterfall_Cliff()
    

#introduction of stage_2 - Transportation Shed
def transportation_Shed():
    print("""
    You safely landed in the river and swam out to the forest.\n
    While you were walking around trying to figure out where you are, you see a 
    light coming from a shelter. As you walk towards the light, you realize it is 
    a Transportation Shed that contains vehicles and weapons. As you draw closer 
    to shed, you notice that it is guarded by two armed guards. For you to safely 
    access the shed you will have to wait for them to sleep.\n""")

    #user is asked to access the transportation_shed
    choice = input(prompt="""\nWould you like to access the shed? Yes/No>\n""")
    choice=choice.lower()

    #results of agreeing to access the shed
    if "yes" in choice or "y" in choice:
        print("Guards will sleep in the next 10 seconds\n")
        time.sleep(10)
        print("Guards are now deep in sleep. Go in!!\n")
        inside_shed()
               
    #user taken back to the begin of the transportation shed intro  
    else:
        print("You will need transportation to get to the shrine.Please try again.")
        input("<Press enter to continue>\n")
        transportation_Shed()

#User enters the shed 
def inside_shed():
    print(""" You safely pass through the sleeping guards then enter the shed. 
    As you enter the shed, you are mesmerized by different objects.""")
    input("""<The following is the list of items in the shed;>\n
            Press enter to view the list>\n""")
    
    #List of objects in the shed
    shed_list = ['helicopter','fire_grenades','machine_Guns','rope']
    print(shed_list)
    
    
    #user is only given 3 items from the shed_list
    
    print("You will only be able to get the following objects from the list")
    
    #for loop for selection of the items in the shed
    for x in shed_list:
        if x != "rope":
            print(x)
    
    #Continuation of the collection of the objects in the shed    
    print("""
        As you pick up the machine_Guns, you trip and fall and the guns fall
        with you and makes a loud noise which wakes the guards up. You get 
        into the helicopter that flies it's self and you land in Jaguar Shrine 
        """)
    jaguar_shrine()

#introduction of the final stage_3
def jaguar_shrine():
    print("""The Jaguar Shrine is a giant mountain in the shape of a Jaguar. 
      It surrounded by Jaguars to defend their territory against the 
      trespassers. As you walk closer to the mountain, suddenly the jewel 
      in your pocket ignites, creating guided pathway towards the Jaguar’s 
      altar where you will have to put back the eye.

      As you begin your journey on the pathway, you see a group of jaguars 
      facing you ready to attack.""")

    
    #user is using the objects from the previous list
    weapon = input(prompt="Which of the weapons retrived from the shed would you like to use to attack back?")
    weapon_list = ['fire_grenades','machine_Guns'] 
    
    #user selects the first item in the list to attack
    if weapon_list[0] in weapon:
        
        print(f"Good choice!! {weapon} wont be able to kill all the jaquars")
        jaguar_eye()
        
    #user selects the first item in the list to attack and fails leads to game over
    elif weapon_list[1] in weapon:
        print(f"Bad choice!! {weapon} wont be able to kill all the jaquars")
        fail()  
        
    #incorrect selection and is taken back to the start 
    else:
        print("You dont have that kind of weapon. Please try again")
        input("Press enter to continue\n")
        start() 

#the winning function
        
def jaguar_eye():
    print("Congrats! You won")
    input(prompt="Type Jumanji to exit the game \n")
    exit()
        
        
           
    
def fail():
    print("""\nYou have failed to lift the curse of Jumanji.""")
    
   


start()
```
